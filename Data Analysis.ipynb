```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.plotting import plot_pca_correlation_graph
```

```python
#Importing the dataset
dataset=pd.read_csv('CLC_train.csv')

Level_dict={'Very low':1,'Low':2,'Moderate':3,'High':4,'Very High':5}
dataset['Level_Ordinal']=dataset.CO_level.map(Level_dict)


X=dataset.iloc[:,2:15].values
y=dataset.iloc[:,16].values
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_t=sc_X.fit_transform(X)
```

```python
#Applying PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X_t=pca.fit_transform(X_t)
explained_variance=pca.explained_variance_ratio_
```

```python
#Visualising the PCA plot
from matplotlib.colors import ListedColormap
X_set,y_set=X_t,y
X1,X2=np.meshgrid(np.arange(start=X_set[:,0].min()-1,stop=X_set[:,0].max()+1,step=0.01),np.arange(start=X_set[:,1].min()-1,stop=X_set[:,1].max()+1,step=0.01))
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set==j,0],X_set[y_set==j,1],c=ListedColormap(('red','green','blue','cyan','magenta'))(i),label=j)
plt.title('PCA Analysis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

```python
#Visualising the correlation circle
feature_names=['CO_GT','PT08_S1','NMHC_GT','C6H6_GT','PT08_S2','NOX_GT','PT08_S3','NO2_GT','PT08_S4','PT08_S5','T','RH','AH']
plot_pca_correlation_graph(X,feature_names)
```
